```python
def convertToLowerCase(text):
  textLower = text.lower()
  return textLower

def performTokenization(text):
  tokens = word_tokenize(text)
  return tokens

def removeStopword(text):
  stop_words = set(stopwords.words('english'))
  res = [t for t in text if not t in stop_words]
  return res

def removePunctuation(text):
  res = [t for t in text if not all(char in string.punctuation for char in t)]
  return res

def removeBlankSpace(text):
   res = [t for t in text if t.strip()!='']
   return res
```

```python
def preprocess(text,show):

  soup = BeautifulSoup(text, 'html.parser')
  text = soup.get_text()

  textLower = convertToLowerCase(text)
  tokens = performTokenization(textLower)
  remStopword = removeStopword(tokens)
  renPunc = removePunctuation(remStopword)
  remBlank = removeBlankSpace(renPunc)

  if show == 'yes':
    print(f"Actual Content:\n{text}\n\n")
    print(f"a. Lowercased Content:\n{textLower}\n")
    print(f"b. Tokenized Content:\n{tokens}\n")
    print(f"c. Without Stopwords:\n{remStopword}\n")
    print(f"d. Without Punctiation:\n{renPunc}\n")
    print(f"e. Without Blank:\n{remBlank}\n")

  return remBlank;
```

```python
def preprocessFile(folder_path,output_path):
    text_files = [f for f in os.listdir(folder_path) if f.endswith('.txt')]

    for file_name in text_files:
        file_path = os.path.join(folder_path, file_name)
        with open(file_path, 'r', encoding='utf-8') as file:
          content = file.read()


        processed_content = preprocess(content,'no')
        processed_text = ' '.join(processed_content)


        output_file_path = os.path.join(output_path, file_name)
        with open(output_file_path, 'w', encoding='utf-8') as file:
          file.write(processed_text)
          file.close()



```

```python
def create_unigram_inverted_index(processedFolder):
    unigram_inverted_index = SortedDict()

    for i in range(len(processed_file_dictionary)):
        file_path = os.path.join(processedFolder, processed_file_dictionary[i])
        with open(file_path, 'r', encoding='utf-8') as file:
          content = file.read()
          file.close()
        words_set = set(content.split())
        sorted_set = sorted(words_set)
        for token in sorted_set:
            if token in unigram_inverted_index:
                unigram_inverted_index[token][0] += 1
                unigram_inverted_index[token][1].add(i)
            else:
                unigram_inverted_index[token] = [1, SortedSet([i])]

    for term, info in unigram_inverted_index.items():
        print(f"Term: {term}, DF: {info[0]}, Document IDs: {info[1]}")

    pickle_file = open('unigram_inverted_index_pkl', 'wb')
    pickle.dump(unigram_inverted_index, pickle_file)
    pickle_file.close()
```

```python
def performAND(l1, l2):
    result = SortedSet()
    i = 0
    j =0
    num_of_operations = 0
    while(i<len(l1) and j<len(l2)):
        if l1[i]==l2[j]:
            result.add(l1[i])
            i+=1
            j+=1
        elif l1[i]<l2[j]:
            i+=1
        else:
            j+=1

    return result


def performOR(l1 , l2):
    res = SortedSet()
    i =0
    j =0
    num_of_operations = 0
    while(i<len(l1)):
        res.add(l1[i])
        i+=1
    while(j<len(l2)):
        res.add(l2[j])
        j+=1
    return res



def performORNOT(l1, l2):
    res = SortedSet()
    for i in range(len(processed_file_dictionary)):
      if i not in l2:
       res.add(i)
    for i in range(len(l1)):
      res.add(l1[i])
    return res


def performANDNOT(l1,l2):
  res = SortedSet()
  for i in l1:
    if i not in l2:
      res.add(i)
  return res


```

```python
def loadPkl(pkl_file_path):
   pkl_file = open(pkl_file_path, 'rb')
   unigram_inverted_index = pickle.load(pkl_file)
   pkl_file.close()
   return unigram_inverted_index
```

```python
def loadQ2():

    unigram_inverted_index = loadPkl('unigram_inverted_index_pkl')
    N = int(input())
    for j in range(N):
        query = input()
        operator = input()
        query = preprocess(query,'no')
        operators = [op.lower().strip() for op in operator.split(',')]
        if(len(query)!=len(operators)+1):
            print('invalid query')
            continue
        index = 0

        if query[index] in unigram_inverted_index:
          firstList = unigram_inverted_index[query[index]][1]
        else:
          firstList = SortedSet()


        for i in range(len(operators)):
          if query[index+1] in unigram_inverted_index:
            second_list = unigram_inverted_index[query[index+1]][1]
          else:
              second_list = SortedSet()

          operation = operators[i]
          if operation == 'and':
            res = performAND(firstList, second_list)
          elif operation == 'or':
            res = performOR(firstList,second_list)
          elif operation=='and not':
            res = performANDNOT(firstList, second_list)
          elif operation =='or not':
            res = performORNOT(firstList, second_list)
          firstList = res
          index = index + 1

        result = [processed_file_dictionary[res_doc] for res_doc in firstList]
        max_len = max(len(query), len(operators))
        print(f'Query {j+1}: ')
        for i in range(max_len):
          if i < len(query):
            print(query[i], end=' ')
          if i < len(operators):
            print(operators[i], end=' ')


        print(f'\nNumber of documents retrieved for query {j+1}: {len(firstList)}')
        print(f'Names of the documents retrieved for query {j+1}: {result}')

```

```python
def create_positional_index(processedFolder):
    positional_index = {}


    for doc_ID, path in enumerate(processed_file_dictionary):
        file_path = os.path.join(processedFolder, processed_file_dictionary[doc_ID])
        with open(file_path, 'r', encoding='utf-8', errors='ignore') as file:
            content = file.read()
            content = content.split()
        tokens = [f for f in content]

        for position, word in enumerate(tokens):
            if word not in positional_index:
                positional_index[word] = {}

            if doc_ID not in positional_index[word]:
                positional_index[word][doc_ID] = []

            positional_index[word][doc_ID].append(position)
    pkl_file = open('positional_index_pkl', 'wb')
    pickle.dump(positional_index, pkl_file)
    pkl_file.close()
    return positional_index
```

```python
def loadQ3():

    positional_index = loadPkl('positional_index_pkl')
    N = int(input())
    for j in range(N):
        query = input()
        query = preprocess(query,'no')
        if not query:
            return []
        initial_docs = positional_index.get(query[0], {})
        valid_docs = []
        for doc_id, positions in initial_docs.items():
            match = True
            for i, word in enumerate(query[1:], 1):
                if doc_id not in positional_index.get(word, {}) or not any(pos - i == position for pos in positional_index[word][doc_id] for position in positions):
                    match = False
                    break
            if match:
                valid_docs.append(processed_file_dictionary[doc_id])
        print(f'Number of documents retrieved for query {j+1} using positional index: {len(valid_docs)}')
        print(f'Number of documents retrieved for query {j+1} using positional index: {valid_docs}')
```
